```python
N = 20
signal_sin = [int(math.sin(i / N * math.pi * 2) * 127) for i in range(N)]
signal_cos = [int(math.cos(i / N * math.pi * 2) * 127) for i in range(N)]
signal_sin
```

```python
def noise(gitter):
    return random.gauss(10, gitter)

def generate_signal(gitter = 0):
    while 1:
        value = yield noise(gitter)
        if value:
            for v in signal_sin:
                yield noise(gitter) + v * value
```

```python
def detect_signal(signal_stream):
    v_sin = 0
    v_cos = 0
    for s_sin, s_cos in zip(signal_sin, signal_cos):
        stream_value = next(signal_stream)
        v_sin += stream_value * s_sin
        v_cos += stream_value * s_cos
    amplitude = (v_sin * v_sin + v_cos * v_cos)
    return amplitude
```

```python
SIGNAL_LENGTH = 100

def rnd(v):
    if v == 0:
        return v
    if v < 0:
        return -int(abs(v) + 0.5)
    return int(v + 0.5)

def sin_cos_1(N):
    import math
    N0 = N
    if N & 1:
        m = 2
        p = 1 / N
        # could be that rounding errors are no problem since we have such a phase
        s_sin = [rnd(math.sin((i + p) / N * math.pi * 2 * m) * 127) for i in range(N)]
        s_cos = [rnd(math.cos((i + p) / N * math.pi * 2 * m) * 127) for i in range(N)]
    else:
        m = 1
        p = 0
        s_sin = [rnd(math.sin(i / N * math.pi * 2) * 127) for i in range(N // 2)]
        s_sin += [-v for v in s_sin]
        #print([rnd(math.cos(i / N * math.pi * 2) * 127) for i in range(N)])
        s_cos = [rnd(math.cos(i / N * math.pi * 2) * 127) for i in range(N // 2)]
        s_cos += [-v for v in s_cos]
        #print(s_cos)
    # there is no problem if this does not add up to zero
    assert sum(s_sin) == 0, "sum(s_sin) == sum({}) == {} for N = {}".format(s_sin, sum(s_sin), N0)
    assert sum(s_cos) == 0, "sum(s_cos) == sum({}) == {} for N = {}".format(s_cos, sum(s_cos), N0)
    return s_sin, s_cos

def sin_cos_2(N):
    N0 = N
    if N & 1:
        N *= 2
        m = 2
        p = 1 / N
    else:
        m = 1
        p = 0
    s_sin = [rnd(math.sin((i + p) / N * math.pi * 2 * m) * 127) for i in range(N)]
    s_cos = [rnd(math.cos((i + p) / N * math.pi * 2 * m) * 127) for i in range(N)]
    #assert sum(s_sin) == 0, "sum(s_sin) == sum({}) == {} for N = {}".format(s_sin, sum(s_sin), N0)
    #assert sum(s_cos) == 0, "sum(s_cos) == sum({}) == {} for N = {}".format(s_cos, sum(s_cos), N0)
    return s_sin, s_cos
    
def sin_cos_3(N):
    s_sin = [int(math.sin(i / N * math.pi * 2) * 127) for i in range(N)]
    s_cos = [int(math.cos(i / N * math.pi * 2) * 127) for i in range(N)]
    return s_sin, s_cos

def sin_cos_4(N):
    # this is used by the positioning system
    import math
    N0 = N
    p = 1 / N
    s_sin = [rnd(math.sin((i) / N * math.pi * 2) * 127) for i in range(N)]
    s_cos = [rnd(math.cos((i) / N * math.pi * 2) * 127) for i in range(N)]
    # there is no problem if this does not add up to zero
    assert sum(s_sin) == 0, "sum(s_sin) == sum({}) == {} for N = {}".format(s_sin, sum(s_sin), N0)
    assert sum(s_cos) == 0, "sum(s_cos) == sum({}) == {} for N = {}".format(s_cos, sum(s_cos), N0)
    return s_sin, s_cos


#N1 = 30
#N2 = 40
#N3 = 50
sin_cos = sin_cos_1
#sin_cos_4(5)
```

```python
def noise(gitter):
    return random.gauss(0, gitter)

 # number of full cycles

def generate_signal(signals, gitter = 0, samples_number = 1000):
    s = list(signals)
    s.sort(key = lambda t: t[1])
    add = []
    samples = []
    for i in range(samples_number):
        sample = int(noise(gitter))
        while s and i == s[0][1]:
            signal = sin_cos(s.pop(0)[0])[0]
            signal = (signal * (SIGNAL_LENGTH // len(signal) + 1))
            assert len(signal) > SIGNAL_LENGTH
            signal = signal[:SIGNAL_LENGTH]
            add.append(signal)
        for a in add[:]:
            sample += a.pop(0)
            if not a: 
                add.remove(a)
        samples.append(sample)
    return samples
                
plt.plot(generate_signal([[30, 30], [50, 100]], 0))
```

```python
def detect_signal_stream(samples, N):
    signal_sin, signal_cos = sin_cos(N)
    assert len(signal_sin) == len(signal_cos)
    processed_samples = [0] * SIGNAL_LENGTH
    v_sin = 0
    v_cos = 0
    processed_i = 0
    wave_i = 0
    amplitudes = []
    for i, sample in enumerate(samples):
        wave_i = i % len(signal_sin)
        processed_wave_i = (i - SIGNAL_LENGTH) % len(signal_sin)
        processed_i = i % SIGNAL_LENGTH
        last_sample = processed_samples[processed_i]
        v_sin += sample * signal_sin[wave_i] - last_sample * signal_sin[processed_wave_i]
        v_cos += sample * signal_cos[wave_i] - last_sample * signal_cos[processed_wave_i]
        #print(v_sin, v_sin, processed_samples)
        amplitude = v_sin * v_sin + v_cos * v_cos
        processed_samples[processed_i] = sample
        amplitudes.append(amplitude)
    return amplitudes


f = 3
if sum(sin_cos(f)[1]) != 0:
    print("cosinus dows not sum up to zero")
samples = generate_signal([[f, 11]], 0, 300)
plt.plot(same_value(samples)) # blue
for sin_cos in (sin_cos_3, sin_cos_2, sin_cos_1):
    ds = detect_signal_stream(samples, f)
    plt.plot(same_value(ds)) # green, red, cyan
    print("{}: last: {} max: {}".format(sin_cos.__name__, ds[-1], max(ds)))
detect_signal_stream(samples, f)[-1]
```

```python
use_average_to_find_index = False

def find_index(amplitudes, width = SIGNAL_LENGTH * 2):
    max_index = max(range(len(amplitudes)), key = amplitudes.__getitem__)
    if not use_average_to_find_index:
        return max_index, 0
    min_index = max_index - width // 2
    if min_index < 0:
        # take less samples and still the middle
        max_index += min_index
        min_index = 0
    samples = amplitudes[min_index: max_index + width // 2]
    #linear_values = [s ** 0.25 for s in samples] # convert to distance which is linear
    #print(max_index, max_index - width // 2, max_index + width // 2, len(amplitudes), samples)
    sm = sum(samples) # can be zero
    if sm == 0:
        return 0, 0
        print("causing error: {}".format(amplitudes))
    return sum((i + min_index) * v for i, v in enumerate(samples)) / sm, sm
           #sum((i + max_index - width // 2) * v for i, v in enumerate(linear_values)) / sum(linear_values)
```

```python
MAXIMUM_LATENCY = 100
OFFSET_TO_AVOID_F1_DOUBLING = SIGNAL_LENGTH
SIGNAL_SAMPLES = 2001

def find_signals(samples, f1, f2, f3):
    volume1 = detect_signal_stream(samples, f1)
    volume2 = detect_signal_stream(samples, f2)
    volume3 = detect_signal_stream(samples, f3)
    window_size = SIGNAL_SAMPLES + MAXIMUM_LATENCY - OFFSET_TO_AVOID_F1_DOUBLING
    window_index = 0
    offsets1 = []
    offsets2 = []
    offsets3 = []
    while window_index < len(samples):
        print("window_index: {} \twindow_size: {}".format(window_index, window_size))
        window1 = volume1[window_index: window_index + window_size]
        #print("window1: {}".format(window1))
        window2 = volume2[window_index: window_index + window_size]
        #print("window2: {}".format(window2))
        window3 = volume3[window_index: window_index + window_size]
        #print("window3: {}".format(window3))

        # does not 
        f_index1, sample_sum1 = find_index(window1)
        f_index2, sample_sum2 = find_index(window2)
        f_index3, sample_sum3 = find_index(window3)
        
        # TODO: this only works for signals with same amplitude. 
        # we need something better in real life if one frequency influences an other
        max_index = f_index1
        offsets1.append(f_index1)
        offsets2.append(f_index2)
        offsets3.append(f_index3)
        #classify
        # assumption: signals do not overlap
        # adding OFFSET_TO_AVOID_F1_DOUBLING influences signal dt
        if window_index + window_size > len(samples):
            # just required for lists
            # we would fill up and process afterwards
            break
        window_index += int(max_index) + OFFSET_TO_AVOID_F1_DOUBLING
    return offsets1, offsets2, offsets3

o1, o2, o3 = find_signals(samples, f1, f2, f3)

#o1.sort()
#o2.sort()
#o3.sort()

plt.plot(o1) # blue
plt.plot(o2) # green
plt.plot(o3) # red
print("\to1: {}\n\to2: {}\n\to3: {}".format(o1, o2, o3))
```

```python
SAMPLE_FREQUENCY = 10000
MAXIMUM_DT = SIGNAL_SAMPLES / SAMPLE_FREQUENCY / 2

def compute_dt(o1, o2, o3):
    o1 = o1[:]
    o2 = o2[:]
    o3 = o3[:]
    
    o1.sort()
    o2.sort()
    o3.sort()
    
    median1 = o1[(len(o1) + 1) // 2]
    median2 = o2[(len(o2) + 1) // 2]
    median3 = o3[(len(o3) + 1) // 2]
    
    print("median1: {} median2: {} median3: {}".format(median1, median2, median3))
    
    signal_samples_computed = median1 + OFFSET_TO_AVOID_F1_DOUBLING
    # these two values should be about the same
    print("signal_samples: {} signal_samples_computed: {}".format(SIGNAL_SAMPLES, signal_samples_computed))
    
    # B sends o1
    # C sends o2
    # A sende o3
    expected_dp = signal_samples_computed / 3
    print("expected_dp: {}".format(expected_dp))
    dp1 = median2 + OFFSET_TO_AVOID_F1_DOUBLING - expected_dp
    print("dp1 raw: {}".format(dp1))
    if signal_samples_computed - dp1 < dp1:
        # check greater than half of rate
        dp1 = dp1 - signal_samples_computed
    elif signal_samples_computed + dp1 < -dp1:
        # check lower than -half of rate
        dp1 = dp1 + signal_samples_computed
    dp2 = median3 + OFFSET_TO_AVOID_F1_DOUBLING - expected_dp * 2
    print("dp2 raw: {}".format(dp2))
#    dp2 = (dp2 + signal_samples_computed / 2) % signal_samples_computed - signal_samples_computed / 2
    if signal_samples_computed - dp2 < dp2:
        dp2 = dp2 - signal_samples_computed
    elif signal_samples_computed + dp2 < -dp2:
        dp2 = dp2 + signal_samples_computed
    print("dp1: {}".format(dp1))
    print("dp2: {}".format(dp2))
    
    dt1 = dp1 / SAMPLE_FREQUENCY
    dt2 = dp2 / SAMPLE_FREQUENCY
    
    return dt1, dt2

compute_dt(o1, o2, o3)
```

```python
def generate_signal_from_dt(dt1, dt2, f1, f2, f3, gitter = 0, seconds = 1):
    o = 0
    signals = []
    print('samples moved f2: {}'.format(int(SAMPLE_FREQUENCY * dt1)))
    print('samples moved f3: {}'.format(int(SAMPLE_FREQUENCY * dt2)))
    for i in range(seconds * SAMPLE_FREQUENCY // SIGNAL_SAMPLES + 1):
        for f, dt in ((f1, 0), (f2, dt1), (f3, dt2)):
            offset = int(o + SAMPLE_FREQUENCY * dt)
            if offset >= 0:
                signals.append((f, offset))
                #print("next offset for {}: {}".format(f, offset))
            o += SIGNAL_SAMPLES // 3
    return generate_signal(signals, gitter, seconds * SAMPLE_FREQUENCY)

plt.plot(generate_signal_from_dt(0,0, 30, 38, 48))
```

```python
def test_dt(dt1, dt2, gitter = 0):
    global o1, o2, o3, samples
    f1, f2, f3 = 3, 4, 5
    print('maximum dt: {}'.format(MAXIMUM_DT))

    # this is a limitation of the sample creation and not of the dt algorithm
    dt1 = int(dt1 * SAMPLE_FREQUENCY) / SAMPLE_FREQUENCY
    dt2 = int(dt2 * SAMPLE_FREQUENCY) / SAMPLE_FREQUENCY
    
    # you can not get outside the possibilities
    dt1 = (dt1 + MAXIMUM_DT) % (MAXIMUM_DT * 2) - MAXIMUM_DT
    dt2 = (dt2 + MAXIMUM_DT) % (MAXIMUM_DT * 2) - MAXIMUM_DT
    
    print("f1: {} f2: {} f3: {}".format(f1, f2, f3))
    samples    = generate_signal_from_dt(dt1,dt2, f1, f2, f3, gitter, 2)#[:4000]
    plt.plot(same_value(samples))

    d1 = detect_signal_stream(samples, f1)
    d2 = detect_signal_stream(samples, f2) # 30 * 2 ** 0.33333333
    d3 = detect_signal_stream(samples, f3) # 30 * 2 ** 0.66666666

    plt.plot(same_value(d1)) # green
    plt.plot(same_value(d2)) # red
    plt.plot(same_value(d3)) # cyan

    o1, o2, o3 = find_signals(samples, f1,f2,f3)
    o1.sort(); o2.sort(); o3.sort()
    print("\to1: {}\n\to2: {}\n\to3: {}".format(o1, o2, o3))
    dt1_, dt2_ = compute_dt(o1, o2, o3)
    print("dt1:  {} dt2:  {}".format(dt1, dt2))
    print("dt1': {} dt2': {}".format(dt1_, dt2_))
    print("error1: {}cm error2: {}cm".format(abs(dt1 - dt1_) * 34320, abs(dt2 - dt2_) * 34320))

use_average_to_find_index = False
test_dt(0.001, 0, 0)
```
